```python
# Divide the dataset into independent and dependent

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
X.shape
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # compute mean and SD
```

```python
scaler.transform(X_train)    # applies z-score
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
# Model training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
# prediction:
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)

mae = mean_absolute_error(y_test,y_pred)
print(mae)

print(np.sqrt(mse))
```

```python
## Accuracy

from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
score
```

```python
# adjusted r-sq
1-(1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=20.0)
```

```python
ridge.fit(X_train,y_train)
```

```python
y_pred = ridge.predict(X_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)

mae = mean_absolute_error(y_test,y_pred)
print(mae)

print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=20.0)
```

```python
lasso.fit(X_train,y_train)
```

```python
y_pred = lasso.predict(X_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)

mae = mean_absolute_error(y_test,y_pred)
print(mae)

print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_net = ElasticNet(alpha=20.0)
elastic_net.fit(X_train,y_train)
```

```python
y_pred = elastic_net.predict(X_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test,y_pred)
print(mse)

mae = mean_absolute_error(y_test,y_pred)
print(mae)

print(np.sqrt(mse))
```
